23. Scrivi un programma che verifichi se un numero è maggiore di un altro e stampi il
risultato.

```python
def major_minor(n):
    if n[0] > n[1]:
        print(f"{n[0]} è maggiore di {n[1]}")
    elif n[0] < n[1]:
        print(f"{n[1]} è maggiore di {n[0]}")
    else:
        print("I due numeri sono uguali")
numeri = [15 , 24]
major_minor(numeri)
numeri2 = [23 , 12]
major_minor(numeri2)
numeri3 = [39 , 39]
major_minor(numeri3)
```

25. Scrivi un programma che confronti due numeri e stampi quello minore.

```python
def minor_number(a, b):
    if a < b:
        print(f"{a} è minore di {b}")
    elif a > b:
        print(f"{b} è minore di {a}")
    else:
        print("I due numeri sono uguali")
minor_number(15, 224)
minor_number(356, 13)
minor_number(123, 123)
```

32. Scrivi un programma che, dati due numeri, stampi:
la loro somma se il primo è maggiore del secondo
la loro differenza se il primo è minore
un messaggio esplicativo se sono uguali

```python
def diff_num(a, b):
    if a > b:
        print(f"{a+b}")
    elif a < b:
        print(f"{a-b}")
    else:
        print("Sono uguali")
diff_num(17, 15)
diff_num(38, 64)
diff_num(15, 15)
```
