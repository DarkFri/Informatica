## Mini-ripasso: tre modi di ciclare
In questa esercitazione verranno usati tre stili. Non sono equivalenti, **occorre usarli” in scenari diversi**.

1. `for i in range(N): ...`  → quando occorre un **contatore** o si vuole lavorare per **posizioni**.
2. `for x in lista: ...`  → quando occorre usare esplicitamente **solo i valori**.
3. `for i, x in enumerate(lista): ...`  → quando occorrono **indice + valore** insieme.

## 2 — Sommare le posizioni pari di una lista
**Consegna**: data `L = [10, 20, 30, 40, 50, 60]`, si calcoli la somma degli elementi in posizione 0,2,4,...

**Vincolo**: usare `for` con `range` (qui il criterio riguarda gli indici).

Risultato atteso con la `L` di default: `10 + 30 + 50 = 90`.


````python
# Usare la lista L definita nella cella Setup (o ridefinirla qui)
somma = 0
# TODO: scorrere gli indici pari e accumulare nella variabile somma
for i in range(0, len(L), 2):
    somma += L[i]
print("Somma posizioni pari:", somma)
````

## Esercizio guidato 3 — Contare quanti valori sono > 100
**Consegna**: data una lista `prices`, contare quanti elementi sono maggiori di 100.

**Vincolo**: iterare direttamente sui valori (`for price in prices`).


````python
count = 0
# TODO: incrementare count per ogni prezzo > 100
for price in prices:
    if price > 100:
        count += 1
print("Prezzi > 100:", count)
````

## 4 — Lunghezza media delle parole
**Consegna**: data `words = ["casa", "programma", "python", "ai"]`, calcolare la lunghezza media.

**Vincolo**: iterare sui valori (qui ti servono solo le stringhe).

Suggerimento: somma le lunghezze e poi dividi per il numero di parole.


````python
tot = 0
# TODO: somma le lunghezze delle parole
avg = 0  # TODO: calcola la media
for word in words:
    tot += len(word)
    avg = tot / len(words)
print("Lunghezza media:", avg)
````

## Esercizio guidato 5 — Stampare “posizione: valore”


**Vincolo**: usare `enumerate`.

## Esercizio guidato 6 — Prima occorrenza di un valore
**Consegna**: data `L = [5, 3, 9, 3, 1]` e `target = 3`, stampare l’indice della **prima** occorrenza.

**Vincolo**: usare `enumerate`.

Nota: quando si trova la prima occorrenza, si può interrompere il ciclo con `break`.


````python
L2 = [5, 3, 9, 3, 1]
target = 3
for i, numero in enumerate(L2):
    if numero == target:
        print(target)
# TODO: trova e stampa l'indice della prima occorrenza di target
````
