### 3. Somma dei valori

```python
prezzi = [10, 20, 15, 30]
def summarize(l):
    result = 0
    lenght_list = len(l)
    for index in range (lenght_list):
        result = result + l[index]
    return result
summarize(prezzi)
```

### 1. Chiedere all'utente un intero e continuare a chiederlo finché l'input non è un numero e non è maggiore di 0; quando è valido stamparlo.

```python
while True:
    try:
        numero = int(input("Inserire un numero"))
        if numero > 0:
            print(numero)
            break
        else:
            print("Errore: il numero deve essere minore di 0")
    except ValueError:
        print("Errore: non puoi inserire un testo")
```

### 2. Chiedere una password e ripetere la richiesta finché l’utente non inserisce esattamente la stringa "python"; alla fine stampare un messaggio di accesso riuscito.

```python
password = ""
while password != "python":
    password = input("Inserisci la password: ")
    if password != "python":
        print("Password errata. Riprova!")
    else:
        print("Accesso avvenuto con successo")
```

### 3. Chiedere all'utente di inserire numeri interi e continuare a chiederlo finché inserisce 0; contare quanti numeri validi sono stati inseriti ed escludere lo 0 finale dal conteggio.

```python
count = 0
while True:
    numero = int(input("Inserisci un numero intero (Inserire 0 per terminare il processo): "))
    if numero == 0:
        break
    count += 1
print(f"Hai inserito {count} numeri prima dello zero")
```

### 4. Chiedere all'utente di inserire numeri interi e continuare a chiederlo finché non inserisce 0; calcolare la somma dei numeri inseriti ed escludere lo 0 finale.

```python
sum = 0
while True:
    numero = int(input("Inserire un numero intero (Inserire 0 per terminare il processo): "))
    if numero == 0:
        break
    sum += numero
    print(numero, sum)
print(f"La somma dei numeri inseriti prima di 0 è: {sum}")
```

### 5. Chiedere all'utente di inserire numeri interi e continuare a chiederlo finché inserisce 0; calcolare la media usando somma e contatore e gestire il caso in cui l’utente inserisca subito 0 stampando un messaggio dedicato.

```python
count = 0
sum = 0
while True:
    numero = int(input("Inserire un numero intero (Inserire 0 per terminare il processo): "))
    if numero == 0:
        break
    count += 1
    sum += numero
if count == 0:
    print("Inserire almeno un numero")
else:
    avg = sum / count
    print(f"Hai inserito: {count} numeri, la loro somma è: {sum}, e la loro media è: {avg}")
```

### 8. Chiedere all'utente di inserire un intero positivo n e stampare con un while tutti i numeri pari da 0 a n controllando la parità con l’operatore %.

### 9. Chiedere all'utente di inserire un intero positivo n e calcolare n! (fattoriale) con un ciclo while inizializzando fattoriale a 1 e moltiplicando progressivamente per i da 1 a n.
                                     n!=n x (n - 1) x (n - 2) x ... 2 x 1
e per defizione:
                                                    0!=1

```python
num = int(input("Inserire un numero positivo per calcolare il fattoriale:"))
fattoriale = 1
n = 1
while n <= num:
    fattoriale = fattoriale * n
    print(f"{n} * {n + 1} = {fattoriale}")
    n += 1
print(f"Il Fattoriale di {num} è: {fattoriale}")
```

### 10. Chiedere all'utente di inserire un intero positivo n e calcolare la somma da 1 a n con un while aggiornando una variabile sum e un contatore i.

```python
num = int(input("Inserire un numero intero massimo:"))
sum = 0
i = 1
conteggio = 0
while i <= num:
    sum = sum + i
    print(f"{conteggio} + {i} = {sum}")
    i += 1
    conteggio += 1
print(f"Da 1 a {num} ci sono {conteggio} numeri e la loro somma è: {sum}")
```
