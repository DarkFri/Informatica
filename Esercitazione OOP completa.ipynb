### 6. Classe Phone . Creare una classe Phone con gli attributi brand , model e battery_level . Inserire un metodo show_battery() che stampi il livello della batteria, inserire un metodo charge che prenda in ingresso un tempo (secondi) che permetta di aggiungere il 5% di carica ogni 2 minuti (60 secondi).

```python
class Phone:
    def __init__(self, model, battery_level):
        self.brand = "Samsung"
        self.model = model
        self.battery_level = max(0, min(100, battery_level))

    def show_battery(self):
        print(f"Livello batteria attutale: {self.battery_level}%")

    def charge_battery(self, seconds):
        charge = (seconds / 120) * 5
        self.battery_level += charge
        if self.battery_level > 100:
            self.battery_level = 100

Federix_Telefix = Phone("S25 Ultra", 72)
print(Federix_Telefix.battery_level)
Federix_Telefix.charge_battery(480)
print(Federix_Telefix.battery_level)
Federix_Telefix.charge_battery(60)
Federix_Telefix.show_battery()
Federix_Telefix.charge_battery(360)
Federix_Telefix.show_battery()
```

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def change_grade(self, new_grade):
        self.grade = new_grade

    def is_passed(self):
        if self.grade >= 6:
            return True
        else:
            return False

Antonio = Student("Antonio", 18, 10)
Antonio.is_passed()
Antonio.change_grade(5)
Antonio.is_passed()
```

### 8. Classe BankAccount con deposito e prelievo. Creare una classe BankAccount con owner e balance . Aggiungere i metodi deposit(amount) e withdraw(amount) . Il prelievo deve avvenire solo se il saldo è sufficiente.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def show_balance(self):
        print("--- Saldo Attuale ---")
        print(f"Proprietario: {self.owner}")
        print(f"Bilancio: {self.balance}")

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance < amount:
            print("Saldo insufficiente")
        else:
            self.balance -= amount
            print("Prelievo avvenuto con successo")

Alessia_Money = BankAccount("Alessia", 10000)
Alessia_Money.deposit(1350)
Alessia_Money.show_balance()
Alessia_Money.withdraw(12000)
Alessia_Money.withdraw(10000)
Alessia_Money.show_balance()
```

### 10. Classe Counter . Creare una classe Counter con l’attributo value . Aggiungere i metodi increment() , decrement() e reset() .

```python
class Counter:
    def __init__(self):
        self.value = 0

    def increment(self, amount):
        self.value += amount

    def decrement(self, amount):
        if self.value < amount:
            return("Impossibile eseguire l'azione")
        else:
            self.value -= amount
        return(self.value)
    def reset(self):
        self.value = 0

Sberlo = Counter()
print(Sberlo.value)
Sberlo.increment(2400)
print(Sberlo.value)
Sberlo.decrement(2200)
print(Sberlo.value)
Sberlo.decrement(400)
print(Sberlo.value)
Sberlo.decrement(300)
print(Sberlo.value)
```

### 11. Classe Book con prestito. Creare una classe Book con title , author , pages e available . Il libro all’inizio è disponibile. Aggiungere i metodi borrow_book() e return_book() . Il prestito può avvenire solo se il libro è disponibile.

```python
class Book:
    def __init__(self, title, author, pages: int, available: bool):
        self.title = title
        self.author = author
        self.pages = pages
        self.available = available

    def borrow_book(self):
        if self.available == True:
            self.available = False
        else:
            return("Non disponibile")

    def return_book(self):
        self.available = True

Forse = Book("Forse un giorno", "Colleen Hoover", 301, True)
Forse.borrow_book()
print(Forse.available)
Forse.borrow_book()
Forse.return_book()
print(Forse.available)
```

### 12. Classe Product con valore totale del magazzino. Creare una classe Product con name , price e quantity . Aggiungere i metodi sell(amount) , restock(amount) e total_value() che restituisca price * quantity . La vendita non deve permettere di andare sotto zero.

```python
class Product:
    def __init__(self, name, price: int, quantity: int):
        self.name = name
        self.price = price
        self.quantity = quantity

    def sell(self, amount):
        if self.quantity < amount:
            print("Prodotto non disponibile")
        else:
            self.quantity -= amount
            print(self.quantity)

    def restock(self, amount):
        self.quantity += amount

    def total_value(self):
        print(f"Prezzo totale: €{self.quantity * self.price}")

Computer = Product("Computer", 214.45, 26)
Computer.sell(30)
Computer.sell(25)
Computer.restock(15)
print(Computer.quantity)
Computer.total_value()
```

### 13. Classe Student con media simulata semplice. Creare una classe Student con name , grade1 , grade2 e grade3 . Aggiungere un metodo average() che restituisca la media dei tre voti, un metodo is_passed() che controlli se la media è almeno 6, e un metodo show_report() che stampi tutti i dati insieme alla media.

```python
class Student:
    def __init__(self, name, grade1, grade2, grade3):
        self.name = name
        self.grade1 = grade1
        self.grade2 = grade2
        self.grade3 = grade3

    def average(self):
        return(round((self.grade1 + self.grade2 + self.grade3) / 3, 2))

    def is_passed(self):
        if self.average() < 6:
            print("BOCCIATO!!!")
        else:
            print("Promosso, caro")

    def show_report(self):
        print("--- Tabella Voti ---")
        print(f"Voto 1: {self.grade1}")
        print(f"Voto 2: {self.grade2}")
        print(f"Voto 3: {self.grade3}")
        print(self.average())

Manu = Student("Manuela Rasta", 8, 5, 6)
Manu.average()
Manu.is_passed()
Manu.show_report()
```

### 14. Classe BankAccount con operazioni controllate. Creare una classe BankAccount con owner , balance e limit . Il limite rappresenta quanto può essere prelevato in una singola operazione. Aggiungere i metodi deposit(amount) , withdraw(amount) , change_limit(new_limit) e show_data() . Il metodo withdraw(amount) deve eseguire il prelievo solo se il valore è positivo, il valore non supera il limite e il saldo disponibile è sufficiente.

```python
class BankAccount:
    def __init__(self, owner, balance, limit):
        self.owner = owner
        self.balance = balance
        self.limit = limit

    def deposit(self, amount):
        self.balance += amount
        print(f"Hai depositato: €{amount}")
        print(f"Saldo attual: €{self.balance:.2f}")

    def withdraw(self, amount):
        if amount > self.balance or amount > self.limit:
            print("Operazione negata: Saldo insufficiente o limite superato.")
        else:
            self.balance -= amount
            print(f"Hai prelevato: €{amount}")
        print(f"Saldo attuale: €{self.balance:.2f}")

    def change_limit(self, new_limit):
        self.limit = new_limit

    def show_data(self):
        print("--- DATA ---")
        print(f"Proprietario: {self.owner}")
        print(f"Saldo: €{self.balance}")

ManuBank = BankAccount("Manuè", 24500, 2300)
ManuBank.withdraw(4000)
ManuBank.withdraw(2300)
ManuBank.show_data()
ManuBank.deposit(1500)
```
